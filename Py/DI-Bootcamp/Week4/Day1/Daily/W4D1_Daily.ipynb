```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Cities
cities = [
    "Moscow", "London", "Tel Aviv", "Jerusalem", "New York",
    "Toronto", "Odessa", "Prague", "Tokyo",
    "Sydney"
]

# Months
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Generate synthetic temperatures between -5 and 35 °C
temperature_data = np.random.uniform(low=-5, high=35, size=(len(cities), len(months)))

# Create DataFrame
df = pd.DataFrame(data=temperature_data, index=cities, columns=months)

# Show first rows
df.head()
```

```python
# 📌 2. Data Analysis

# Annual average temperature for each city
annual_avg = df.mean(axis=1)

# Find the hottest and coldest city
hottest_city = annual_avg.idxmax()
coldest_city = annual_avg.idxmin()

print("🌡 Annual Average Temperatures (°C):")
print(annual_avg.round(2))

print("\n🔥 Hottest city:", hottest_city, "with", round(annual_avg[hottest_city], 2), "°C")
print("❄️ Coldest city:", coldest_city, "with", round(annual_avg[coldest_city], 2), "°C")
```

```python
# 📌 3. Visualization

plt.figure(figsize=(12,6))

# Plot monthly temperatures for all cities
for city in df.index:
    plt.plot(months, df.loc[city], marker="o", label=city, alpha=0.7)

plt.title("📈 Average Monthly Temperatures by City", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.show()
```
