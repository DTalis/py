```python
#Exercise 2
import numpy as np
data = [12, 15, 13, 12, 18, 20, 22, 21]

mean_np = np.mean(data)
median_np = np.median(data)
variance_np = np.var(data)
std_dev_np = np.std(data)

print("Mean: ", mean_np)
print("Median: ", median_np)
print("Variance: ", variance_np)
print("Standard Deviation: ", std_dev_np)
```

```python
#Exercise 3

import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mean = 50
std_dev = 10

# Generate x values (range around the mean)
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 200)

# Calculate the probability density function (PDF)
y = norm.pdf(x, mean, std_dev)

# Plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label=f"Normal Distribution\nMean={mean}, Std={std_dev}")
plt.title("Normal Distribution using SciPy")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
#Exercise 4
from scipy import stats

# Generate two random samples from normal distributions
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

# Perform independent two-sample t-test
t_stat, p_value = stats.ttest_ind(data1, data2)

# Print results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret result (at 5% significance level)
if p_value < 0.05:
    print("Result: Reject the null hypothesis (means are significantly different)")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference)")
```

```python
#Exercise 5
# Dataset
house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")

# Predict price for house size = 90 mÂ²
predicted_price = slope * 90 + intercept
print(f"Predicted Price for 90 mÂ² house: {predicted_price:.2f}")

# Optional: Plot regression line
import matplotlib.pyplot as plt

plt.scatter(house_sizes, house_prices, label="Data Points")
plt.plot(house_sizes, slope * house_sizes + intercept, color='red', label="Regression Line")
plt.title("Linear Regression: House Size vs Price")
plt.xlabel("House Size (mÂ²)")
plt.ylabel("House Price")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Exercise 6
# Dataset
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# Perform one-way ANOVA
f_stat, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print(f"F-value: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret results
if p_value < 0.05:
    print("Result: Reject the null hypothesis â†’ Fertilizers have significantly different effects.")
else:
    print("Result: Fail to reject the null hypothesis â†’ No significant difference between fertilizers.")
```

```python
#Exersice 7
from scipy.stats import binom

# Parameters
n = 10          # number of trials (flips)
p = 0.5         # probability of success (head)
k = 5           # number of successes

# Calculate probability of exactly 5 heads
prob_5_heads = binom.pmf(k, n, p)
print(f"Probability of getting exactly 5 heads in 10 flips: {prob_5_heads:.4f}")
```

```python
#Exercise 8
import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Sample dataset
data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

# Extract columns
ages = data['age']
incomes = data['income']

# Calculate Pearson correlation (linear relationship)
pearson_corr, pearson_p = pearsonr(ages, incomes)

# Calculate Spearman correlation (rank-based)
spearman_corr, spearman_p = spearmanr(ages, incomes)

# Print results
print(f"Pearson Correlation: {pearson_corr:.4f} (p-value: {pearson_p:.4f})")
print(f"Spearman Correlation: {spearman_corr:.4f} (p-value: {spearman_p:.4f})")
```
