```python
#Exercise 2
import matplotlib.pyplot as plt

temperature = [72, 74, 76, 80, 82, 78, 75]
time = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

plt.plot(time, temperature, marker="o")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°F)")
plt.title("Temperature Over Time")
plt.show()
```

```python
sales = [5000, 5500, 6200, 7000, 5500, 5000, 4900, 4500, 2300, 4300, 3500, 2000]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

plt.bar(months, sales)
plt.xlabel("Month")
plt.ylabel("Sales Amount ($)")
plt.title("Monthly Sales")
plt.show()
```

```python
#Exercise 5
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/Student Mental health.csv")

plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Choose your gender", hue="Do you have Anxiety?", palette="coolwarm")

plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.title("Anxiety Levels Across Genders")
plt.legend(title="Anxiety")
plt.show()
```

```python
#Exercise 6
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/Student Mental health.csv")
print(df.columns.tolist())

# ð Step 3: Convert panic attack responses to numeric values (Yes=1, No=0)
df['Panic_Attacks_Numeric'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})

# Quick check
df[['Age', 'Do you have Panic attack?', 'Panic_Attacks_Numeric']].head()

# ð Step 4: Create scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x='Age',
    y='Panic_Attacks_Numeric',
    hue='Panic_Attacks_Numeric',
    style='Panic_Attacks_Numeric',
    palette={0: "blue", 1: "red"}
)

# ð Step 5: Customize the plot
plt.title("Relationship between Age and Panic Attacks")
plt.xlabel("Age")
plt.ylabel("Panic Attacks (1=Yes, 0=No)")
plt.yticks([0, 1], ["No", "Yes"])
plt.legend(title="Panic Attacks")
plt.grid(True, linestyle="--", alpha=0.6)

# ð Step 6: Display the plot
plt.show()

```
