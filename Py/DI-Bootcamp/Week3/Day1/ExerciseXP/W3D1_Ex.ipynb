What is Data Analysis?

Data analysis is the systematic process of collecting, cleaning, transforming, and interpreting data to uncover meaningful patterns, trends, and insights. It involves applying statistical methods, logical reasoning, and computational tools to raw information so that it can support decision-making. At its core, data analysis converts data from ‚Äúnoise‚Äù into knowledge that organizations and individuals can understand and act upon.

Why is Data Analysis Important in Modern Contexts?

In today‚Äôs world, data is generated at an unprecedented scale‚Äîfrom online transactions and social media activity to machine sensors and business systems. Data analysis has become essential because it helps organizations make informed, evidence-based decisions rather than relying on intuition. It improves efficiency, identifies risks, predicts future outcomes, and uncovers opportunities for innovation. From healthcare to finance, companies depend on data analysis to stay competitive, understand customer behavior, optimize processes, and reduce uncertainties.

Three Key Areas Where Data Analysis Is Applied

1. Business & Marketing
Companies perform data analysis to understand customer behavior, measure campaign effectiveness, forecast sales, and segment audiences. For example, e-commerce platforms use customer data to recommend products, personalize ads, and optimize prices.

2. Healthcare
In healthcare, data analysis helps diagnose diseases earlier, analyze patient records, monitor public health trends, and support medical research. Hospitals use predictive models to estimate patient risks, while pharmaceutical companies analyze clinical trial results to evaluate new drugs.

3. Finance & Risk Management
Banks, fintech companies, and insurance firms heavily rely on data analysis to detect fraud, assess creditworthiness, prevent financial crime, and manage market risks. Machine learning models can predict loan defaults, evaluate transaction anomalies, and support strategic investment decisions.

Qualitative Columns (Categorical variables)

1. Species

	‚Ä¢	Iris-virginica

```python
# ---------------------------------------------
# Exercise: Basic Data Analysis with Iris Dataset
# ---------------------------------------------

# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load Dataset
df3 = pd.read_csv("/content/Iris.csv")

print("First 5 rows of the dataset:")
display(df3.head())

# ---------------------------------------------
# 3. Calculate Mean, Median, Mode for a quantitative column
# ---------------------------------------------
column = "PetalLengthCm"

mean_val = df3[column].mean()
median_val = df3[column].median()
mode_val = df3[column].mode()[0]

print("\nStatistical Analysis for:", column)
print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)

# ---------------------------------------------
# 4. Histogram (Matplotlib)
# ---------------------------------------------
plt.figure(figsize=(8,5))
plt.hist(df3[column], bins=20, edgecolor="black")
plt.title("Histogram of " + column)
plt.xlabel(column)
plt.ylabel("Frequency")
plt.show()

# ---------------------------------------------
# 5. Bar Chart for Species Count
# ---------------------------------------------
species_count = df3["Species"].value_counts()

plt.figure(figsize=(8,5))
plt.bar(species_count.index, species_count.values)
plt.title("Count of Samples per Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()

# ---------------------------------------------
# 6. Seaborn Histogram with KDE
# ---------------------------------------------
plt.figure(figsize=(8,5))
sns.histplot(df3[column], kde=True)
plt.title("Seaborn Histogram with KDE - " + column)
plt.xlabel(column)
plt.show()

# ---------------------------------------------
# 7. Printed Findings (for documentation)
# ---------------------------------------------
print("\nFindings:")
print(f"- The mean petal length = {mean_val:.2f}")
print(f"- The median petal length = {median_val:.2f}")
print(f"- The mode petal length = {mode_val:.2f}")
print("- The histogram shows three clusters of petal lengths representing the three Iris species.")
print("- The bar chart confirms there are 50 samples per species, evenly distributed.")
```
