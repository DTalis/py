1. Introduction to Machine Learning

Machine Learning (ML) is a field of artificial intelligence that enables computers to learn patterns from data and improve their performance on a task without being explicitly programmed.
Instead of using hard-coded rules, ML models automatically identify relationships in datasets and use them to make predictions, classifications, or decisions.

‚≠ê B. Unsupervised Learning

	‚Ä¢	Clustering products into categories
	‚Ä¢	Detecting anomalies without predefined labels
